# O

```python
import time
import matplotlib.pyplot as plt
```

```python
def linear_search(arr, target):
    for i in arr:
        if i == target:
            return True
    return False
```

```python
sizes = [100, 500, 1000, 2000, 3000, 4000, 5000]
times = []

for size in sizes:
    data = list(range(size))
    target = size - 1  # el peor caso es buscar el Ãºltimo elemento

    start = time.time()
    linear_search(data, target)
    end = time.time()

    times.append(end - start)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(sizes, times, marker='o', label="O(n) - BÃºsqueda lineal")
plt.xlabel("TamaÃ±o del arreglo")
plt.ylabel("Tiempo de ejecuciÃ³n (segundos)")
plt.title("Complejidad O(n): BÃºsqueda lineal")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def linear_search(arr, target):
    for i in arr:
        if i == target:
            return True
    return False
```

```python
sizes = [100, 500, 1000, 2000, 3000, 4000, 5000]
times = []
```

```python
for size in sizes:
    data = list(range(size))
    target = size // 2 # promedio: elemento a la mitad

    start= time.time()
    linear_search(data, target)
    end = time.time()

    times.append(end - start)
```
