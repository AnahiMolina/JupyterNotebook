## 1. Vectores

```r
#vector numérico en R
v1<-c(1,2,3,4,5)
print(v1)
class(v1)
length(v1)
```

```r
#vector de caracteres en R
v2<-c("A","B","A","B","B")
print(v2)
class(v2)
length(v2)
```

```r
#vector de valores lógicos (T / F)
v3<-c(TRUE, FALSE, TRUE, TRUE, FALSE)
print(v3)
class(v3)
length(v3)
```

```r
v4<-c(26,"clase",FALSE)
v4
```
