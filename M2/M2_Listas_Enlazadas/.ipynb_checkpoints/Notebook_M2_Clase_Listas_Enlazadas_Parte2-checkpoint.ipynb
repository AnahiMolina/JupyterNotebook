```python
# Python Program to insert the node at the beginning of
# Linked List

class Node:
    def __init__(self, new_data):
        # Initialize the node's data
        self.data = new_data
        # Set the next pointer to None
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None  # Initialize head as None

    def insert_at_front(self, new_data):
        # Create a new node with the given data
        new_node = Node(new_data)

        # Make the next of the new node point to the current head
        new_node.next = self.head
        self.head = new_node  # El nodo recién creado se convierte en cabeza de la lista.

        # Return the new node as the new head of the list
        #return new_node


    def print_list(self):
        # Start from the head of the list
        curr = self.head

        # Traverse the list
        while curr is not None:
            # Print the current node's data
            print(f" {curr.data}", end="")

            # Move to the next node
            curr = curr.next

        # Print a newline at the end
        print()

#Insert Node at the End of a Linked List
    # Given the head of a list and an int, appends
    # a new node at the end and returns the head.
    def append(self, new_data):

        # Create a new node
        new_node = Node(new_data)

        # If the Linked List is empty, make the 
        # new node as the head and return
        if self.head is None:
            return new_node

        # Store the head reference in a temporary variable
        last = self.head

        # Traverse till the last node
        while last.next:
            last = last.next

        # Change the next pointer of the last 
        # node to point to the new node
        last.next = new_node

        # Return the head of the list
        return self.head
    
#Insert a Node At a Specific Position in Linked List

    # function to insert a Node at required position
    def insert_pos(self, pos, data):

        # This condition to check whether the
        # position given is valid or not.
        if pos < 1:
            return self

        # head will change if pos=1
        if pos == 1:
            new_node = Node(data)
            new_node.next = self.head
            size += 1
            return new_node

        curr = self.head

        # Traverse to the node that will be 
        # present just before the new node
        for _ in range(1, pos - 1):
            if curr == None:
                break
            curr = curr.next

        # if position is greater
        # number of nodes
        if curr is None:
            return self.head

        new_node = Node(data)

        # update the next pointers
        new_node.next = curr.next
        curr.next = new_node

        return self.head

### Deletion at beginning (Removal of first node) in a Linked List

# Delete the head node and return the new head
    def delete_head(self):

        # Check if the list is empty
        if self.head is None:
            return None

        # Store the current head in a temporary variable
        temp = self.head

        # Move the head pointer to the next node
        self.head = self.head.next

        # Delete the old head by removing the reference
        del temp

        return self.head
    
# Python program to delete the last node of linked list
    #Function to remove the last node of the linked list
    def remove_last_node(self):
        # If the list is empty, return None
        if not self.head:
            return None

        # If the list has only one node, delete it and return None
        if not self.head.next:
            return None

        # Find the second last node
        second_last = self.head
        while second_last.next.next:
            second_last = second_last.next

        # Delete the last node
        second_last.next = None

        return self.head
    

# Function to delete a node at a given position
    def deleteNode(self, position):
        temp = self.head
        prev = None

        # Base case if linked list is empty
        if temp is None:
            return self.head

        # Case 1: Head is to be deleted
        if position == 1:
            head = temp.next
            return self.head

        # Case 2: Node to be deleted is in middle
        # Traverse till given position
        for i in range(1, position):
            prev = temp
            temp = temp.next
            if temp is None:
                print("Data not present")
                return self.head

        # If given position is found, delete node
        if temp is not None:
            prev.next = temp.next

        return self.head
```

```python
# Iterative Python program to search an element in linked list
def search_key(self, key):
  
    # Initialize curr with the head of linked list
    curr = self.head

    # Iterate over all the nodes
    while curr is not None:

        # If the current node's value is equal to key,
        # return true
        if curr.data == key:
            return True

        # Move to the next node
        curr = curr.next

    # If there is no node with value as key, return false
    return False
```
