```python
import numpy as np
# Crear una clase  class MyStack. 
class MyStack:
    def __init__(self):
        self.stack=[]
    def push (self, data):
        self.stack.append(data)
        print(self.stack)
    def pop (self):
        self.stack.pop()
        print(self.stack)
    def empty (self):
        if len(self.stack)==0: 
            print("Empty Stack")
        else: 
            print("No Empty Stack")
    def size (self):
        return len(self.stack)
    def top (self):
        return(self.stack[-1])
```

```python
# A partir de MyStack crear una pila ùë∫ùüè de tama√±o 30, con n√∫meros aleatorios de una distribuci√≥n normal est√°ndar; usar dos cifras decimales 
# y semilla=2104). 
S1 = MyStack()
s1 = np.around(np.random.default_rng(2104).standard_normal(30),2)
```

```python
for i in range(30):
    S1.push(s1[i])
```

```python
# Verificar la longitud de la pila ùë∫ùüè. 
S1.size()
```

```python
# Identificar el dato en el tope (cima) de la pila ùë∫ùüè. 
S1.top()
```

```python
# Eliminar los √∫ltimos 5 n√∫meros de la pila ùë∫ùüè 
for i in range (5):
    S1.pop()
```

```python
# Verificar la longitud de la pila ùë∫ùüè. 
S1.size()
```

```python
# Eliminar los datos restantes para limpiar la pila ùë∫ùüè. 
for i in range (25):
    S1.pop()
```

```python
# Verificar si la pila ùë∫ùüè est√° vac√≠a.
S1.empty()
```
