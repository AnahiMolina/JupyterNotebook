```python
import numpy as np
class Stack:
    #Constructor
    def __init__(self):
        self.stack=[]
    #Métodos
    def push (self, data):
        self.stack.append(data)
        print(self.stack)
    def pop (self):
        self.stack.pop()
        print(self.stack)
    def empty (self):
        if len(self.stack)==0: 
            print("Empty Stack")
        else: 
            print("No Empty Stack")
    def size (self):
        return len(self.stack)
    def top (self):
        return(self.stack[-1])    
```

#### 1) Crear una pila st1 de tamaño 100, con números aleatorios de una distribución normal estándar (dos cifras decimales y semilla=3103). 

```python
st1 = Stack()
v1=np.around(np.random.default_rng(3103).standard_normal(100),2)
for i in range(100):
    st1.push(v1[i])
```

```python
# 1.1 Verificar la longitud de la pila 𝒔𝒕𝟏. 
st1.size()
```

```python
# 1.2 Identificar el dato en el tope (cima) de la pila 𝒔𝒕𝟏. 
st1.top()
```

```python
# 1.3 Eliminar los últimos 50 números de la pila 𝒔𝒕𝟏 
for i in range(50):
    st1.pop()
```

```python
# 1.4 Verificar la longitud de la pila 𝒔𝒕𝟏. 
st1.size()
```

```python
# 1.5 Identificar el dato en el tope (cima) de la pila 𝒔𝒕𝟏. 
st1.top()
```

```python
# 1.6 Eliminar los datos restantes para limpiar la pila 𝒔𝒕𝟏. 
for i in range(50):
    st1.pop()
```

```python
# 1.7 Verificar si la pila 𝒔𝒕𝟏 está vacía.
st1.empty()
```

#### 3) Crear una pila 𝒔𝒕𝟑 de tamaño 60, donde los datos [A ,B ,C] se repitan hasta llenar la pila. 

```python
st3 = Stack()
v3 = ["A","B","C"]
for i in range (60):
    st3.push(v3[i % len(v3)])
```

```python
# 2.1 Verificar la longitud de la pila 𝒔𝒕𝟑. 
st3.size()
```

```python
# 2.2 Identificar el dato en el tope (cima) de la pila 𝒔𝒕𝟑. 
st3.top()
```

```python
# 2.3 Eliminar los últimos 10 datos de la pila 𝒔𝒕𝟑. 
for i in range(10):
    st3.pop()
```

```python
#2.4 Verificar la longitud de la pila 𝒔𝒕𝟑. 
st3.size()
```
