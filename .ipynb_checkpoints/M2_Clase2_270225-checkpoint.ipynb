Ejercicio: Crear una lista que contenga:
- un vector lógico de longitud 10.
- una matriz dim(5,5) con datos numéricos aleatorios con distribución normal estándar N (media=0, varianza=1).
- un data frame de tres variables con 10 casos cada una.

```r
#Vector lógico
x=rep(c(TRUE, FALSE),5)
x
```

```r
#Matriz numérica
#numeros aleatorios con distriución normal estándar: rnorm(n, mean = 0, sd = 1)
#Datos reproducibles, se agrega semilla

set.seed(2025)
y=round(rnorm(n=25),1) 

M=matrix(data=y, 5, 5)
M
```

```r
#Data frame
#Vectores faltantes
w=c(1:10)
w

z=rep(c("Si","No"), 5)
z

DF=data.frame(x,w,z)
DF
```

```r
#LISTA
L=list(x,M,DF)
L
```

# Características y Operaciones por listas

```r
#Longitud es el número de elementos u objetos de la lista.
length(L)

#identifica el tipo de objeto
class(L)

#Identifica si el objeto es una lista. (Boolean Answer)
is.list(L)
```

```r
#Extraer elementos de una lista en R
#Primer elemento de la lista 
#OBJETO[posición]
#En R las posiciones o índices comienzan en 1

L[1] #Extraer elemento por posición de las listas
```

```r
L[3] #Extrae el elemento número 3
L[5] #Elemento no existe, está fuera del rango de longitud de L
```

```r
#Primer elemento del primer elemento de la lista
#Para este ejercicio se extrae la primera componenete del vector
L[[1]][1]

#Primer elemento del segundo elemento de la lista
#En el ejercicio L[2] es una matriz, extraer la primera fila
L[[2]][1,]

#Extraer de L la primera columna de la matriz guardada en L[2]
L[[2]][,1]
M
```

```r
#Extraer de L la segunda columna del data frame
L[[3]][,2]
L[[3]][2]
```

```r
#Añadir elementos a una lista en R

#Crear tabla
T1<-table(z)
#T1

#Agregar la tabla T1 a la lista L, al final de la lista
L[[4]]<-T1
L
```

```r
#agregar vector z a la lista
L[[1]]<-z
L
```

## Borrar elementos de una lista

```r
#Borrar el vector z de la lista L
#Se pueden poner Null o -n (elemento)
L[-1]
L
```

```r
#Tamaño de la lista actual
length(L)
L
```

```r
#Crear una lista L2 con los vectores x,w,z
L2<-list(x,w,z)
L2
is.list(L2)
```

## Unir listas en R

```r
#Unir  las listas L1 y L2 en una lista L3
#concatenar las listas
L3=append(L,L2)
L3
```

## Comparar dos listas

```r
#Identificar los elememntos que tienen en común L y L2
#Devuelve los elementos comunes
intersect(L, L2)
```

```r
#Elementos diferentes en las listas
#settdif(lista1, lista2)
setdiff(L,L3)
```

```r
#Elementos diferentes en las listas
#settdif(lista1, lista2)
setdiff(L3,L)
```

```r
#Elementos diferentes en las listas
#settdif(lista1, lista2)
setdiff(L,L2)
```
